```python
import seaborn as sns
#seaborn is used for a dataframe.
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
A=sns.load_dataset("penguins")  #dataset fmri
# https://github.com/mwaskom/seaborn-data
```

```python
A.head()
```

# Violin Plot

```python
#violin plot is same as box plot
#shows distn of num var across many categorical var
sns.set(style='darkgrid') #style must be one of white, dark, whitegrid, darkgrid, ticks
sns.violinplot(x='sex',y='bill_length_mm',data=A,linewidth=2)

#linewidth=border size of violin
#palette
#color=for same color for all violin
#order=['Female','Male']
```

```python
sns.violinplot(x='species',y='bill_length_mm',data=A,hue='sex',linewidth=2,split=True)

#for spliting there must be exactly 2 values in hue
#scale= count...
#inner=quart(divides in quatile)(box,point,stick)
```

# Pair Plot

```python
#Used when we draw ML algo, to compare different ML algo

sns.pairplot(A,hue='sex',palette='YlOrBr')
#differentiate according to sex ,therefore hue=sex

```

```python
#for specific graph

# vars=keyword for variables
sns.pairplot(A,hue='sex',vars=['bill_length_mm','bill_depth_mm'],kind='reg')  #kind=hist,scatter
#hue_order=[..]
#x_vars=[..]
#y_vars=[..]
#kind=reg
```

# Strip Plot

```python
# Basically a scatter plot that differentiates different categories
#alternative for box and violin plot
#alpha=for light dark of points
sns.stripplot(x='bill_length_mm',y='bill_depth_mm',data=A)  #only take num valuesbill_length_mm
```

# Box Plot

```python
sns.boxplot(x='island',y='bill_depth_mm',data=A,color='r')

#orient='y'
```

# Factor Plot

```python
#factor plot is renamed as catplot

# can draw different graphs using kind parameter
sns.factorplot(x='island',y='bill_depth_mm',data=A,hue='sex',kind='bar') #kind=bar,box
```

# Cat Plot

```python
#catplot is used to make multiple plots using a single function

sns.catplot(x='island',y='bill_depth_mm',data=A,kind='violin')  #kind=box,bar,violin,countstrip,swarm,boxen,point

#height
```

# FacetGrid

```python
fg=sns.FacetGrid(A)

fg.map(plt.scatter,'bill_length_mm','bill_length_mm')
plt.show()
#take any 2 parameters in map
```

```python
#to get multiple plots
fg=sns.FacetGrid(A,col='sex')

fg.map(plt.scatter,'bill_length_mm','bill_length_mm')
plt.show()
#take any 2 parameters in map
#.add_legend()
```
