# Different Plots in Seaborn

```python
import seaborn as sns
#seaborn is used for a dataframe.
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data1=[1,2,3,4,9,6,7,8]
data2=[5,7,8,3,5,7,8,1]

df=pd.DataFrame({'data1':data1,'data2':data2})
```

```python
plt.plot(data1,data2)
```

```python
sns.lineplot(x=data1,y=data2,data=df)
```

```python
sns.lineplot(x='data1',y='data2',data=df)

# data1=[1,2,3,4,9,6,7,8]
# data2=[5,7,8,3,5,7,8,1]
```

# For importing inbuilt data in seaborn 

```python
A=sns.load_dataset("iris")
# https://github.com/mwaskom/seaborn-data
```

```python
A.head(10)
```

# Line plot

```python
sns.lineplot(x='species',y='sepal_width',data=A,style='sepal_length',palette='Accent',legend=False)
#hue adds another color line (feature)in the same graph #hue='sepal_length'
#style changes lline format (here it is dotted)
#palette gives color
#legend  will show values on error
#if legend is false then scale is not visible 
plt.xlabel('Species')
plt.grid()

plt.ylabel('2nd feature sepal-width')
plt.title('Iris dataset representation')
```

# Histogram 

```python
sns.displot(A['sepal_length'],bins=[4,5,6,7,8],color='red')

#bins is used fro range on x axis
plt.show()
```

# Bar plot

```python
sns.barplot(x='species',y='sepal_length',data=A)
```

```python
# OR

sns.barplot(x=A.species,y=A.sepal_length)

#hue= another feature can be added
#hue_order = to change order of multiple features
#saturation
#palette
#orient
#errcolor=y,r,g...
#errwidth=num
#capsize=num

sns.set(style='darkgrid')   #dark
order=['virginica','setosa','versicolor species']
sns.barplot(x=A.species,y=A.sepal_length,order=order)
```

# Scatter plot

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=A,hue='species',size='species')

#He hue is the differentiating variable(on what basis the data is differentiated)
#size=differentiating variable with different size(we can give our size also)
#marker
#alpha (alpha value is always between 0 and 1)
```

# Count Plot

```python
sns.countplot(x='species',data=A)

#countplot-> counts the no. of records by category
#barplot-> value of each category,plots the 'mean' of variables by category
#in barplot you have to give x and y(2 var)
#in count plot you only have to give x(1 var)

#palette,color,hue,saturation=0to1
```
